# customer churn analysis for e-commerse dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
from pandas_profiling import ProfileReport
```

```python
df_description = pd.read_excel("Dataset.xlsx")
df_description
```

```python
df = pd.read_excel("Dataset.xlsx" , sheet_name = "E Comm")
df.head()
```

```python
report = ProfileReport(df)
report.to_file("profile.html")
```

```python
df.isnull().sum()
```

```python
#fill numeric values with mean
df['Tenure'] = df['Tenure'].fillna(np.mean(df['Tenure']))
```

```python
df['WarehouseToHome'] = df['WarehouseToHome'].fillna(np.mean(df['WarehouseToHome']))
df['HourSpendOnApp'] = df['HourSpendOnApp'].fillna(np.mean(df['HourSpendOnApp']))
df['OrderAmountHikeFromlastYear'] = df['OrderAmountHikeFromlastYear'].fillna(np.mean(df['OrderAmountHikeFromlastYear']))
df['CouponUsed'] = df['CouponUsed'].fillna(np.mean(df['CouponUsed']))
df['OrderCount'] = df['OrderCount'].fillna(np.mean(df['OrderCount']))
df['DaySinceLastOrder'] = df['DaySinceLastOrder'].fillna(np.mean(df['DaySinceLastOrder']))
```

```python
df['Tenure'] = [round(float(int(i)),2) for i in df['Tenure']]
df['WarehouseToHome'] = [round(float(int(i)),2) for i in df['WarehouseToHome']]
df['HourSpendOnApp'] = [round(float(int(i)),2) for i in df['HourSpendOnApp']]
df['OrderAmountHikeFromlastYear'] = [round(float(int(i)),2) for i in df['OrderAmountHikeFromlastYear']]
df['CouponUsed'] = [round(float(int(i)),2) for i in df['CouponUsed']]
df['OrderCount'] = [round(float(int(i)),2) for i in df['OrderCount']]
df['DaySinceLastOrder'] = [round(float(int(i)),2) for i in df['DaySinceLastOrder']]
```

## churn(1 = true , 0 = False), x format(citytier , churn flag)

## above plot shows that city with  tier 1 has the highest churn

```python
df_cleaned[['HourSpendOnApp','CityTier']].iloc[0]
```

```python
#plt.figure(figsize=(10,10))
#mpl.style.use('ggplot')

#plt.bar(df_cleaned['CityTier'],df_cleaned['HourSpendOnApp'])
t1,t2,t3,i = [],[],[],0
while i < len(df_cleaned['CityTier']):
    item = df_cleaned[['HourSpendOnApp','CityTier']].iloc[i]
    if item['CityTier']==1.0:
        t1.append(item['HourSpendOnApp'])
    if item['CityTier']==2.0:
        t2.append(item['HourSpendOnApp'])
    if item['CityTier']==3.0:
        t3.append(item['HourSpendOnApp'])
    i+=1

hoursdata=[np.sum(t1),np.sum(t2),np.sum(t3)]
tiers = ['tier1','tier2','tier3']

plt.figure(figsize=(5,5))
mpl.style.use('ggplot')
plt.bar(tiers,hoursdata)
plt.xlabel('CityTier')
plt.ylabel("Total Hour Spend On App")
plt.title('Total Hour Spend On App in Each CityTier')
```

## total hours spent on app is highest in tier 1 cities

```python
mpl.style.use('seaborn-pastel')
df_cleaned[['CityTier','HourSpendOnApp']].value_counts().plot(kind= 'barh')
plt.title("no of user according to city tier and hours spent on app")
```
